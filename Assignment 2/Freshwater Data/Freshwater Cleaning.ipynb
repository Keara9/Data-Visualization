```python
df = pd.read_csv("NewZealand Freshwater Original.csv")
df
```

```python
df = df.drop(range(9,14))
df
```

```python
df = df.drop(columns=[col for col in df.columns if col.startswith(tuple(str(year) for year in range(1960, 1985)))])
df = df.drop(columns=[col for col in df.columns if col.startswith(tuple(str(year) for year in range(2022, 2024)))])
df
```

```python
df.replace('..', pd.NA, inplace=True)
df
```

```python
#Transform the data from wide to long format
df_long = df.melt(
    id_vars=["Series Name"],  # Keep 'Series Name' as the identifier
    var_name="Year",          # Name the year column
    value_name="Value"        # Name the value column
)

# Extract the year from the 'Year' column (e.g., "1960 [YR1960]" -> 1960)
df_long["Year"] = df_long["Year"].str.extract(r"(\d{4})")

# Drop rows with missing years
df_long = df_long.dropna(subset=["Year"])

# Convert the 'Year' column to integers
df_long["Year"] = df_long["Year"].astype(int)

df = df_long
df
```

```python
missing_values = df.isnull().sum()  
missing_values
```

```python
df.to_csv("NewZealand Freshwater cleaned.csv", index=False)
```
